# Análisis: Que factores afectan el Monto de las Propinas
### DataSet: Tips

```python
# Importaremos 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style='darkgrid')

#Cargamos el DataSet

df = sns.load_dataset('tips')
df
```

```python
sns.set(style='darkgrid')
```

```python
# Gráfico de barra de propinas promedio por sexo
sns.barplot(data=df,x='sex', y='tip', ci=None)
plt.title('Propinas promedio por sexo')
plt.xlabel('Género')
plt.ylabel('Propina')
plt.show()
```

Los hombres en promedio pagan más propina

```python
sns.barplot(data=df, x='day', y='tip', ci=None)
plt.title('Propinas Promedio por día')
plt.xlabel('Día de la semana')
plt.ylabel('Propina')
plt.show()
```

Los domingos se pagan más propinas

```python
sns.barplot(data=df,x='time',y='tip',ci=None)
plt.title('Propinas según servicio')
plt.xlabel('Servicio')
plt.ylabel('Propina')
plt.show()
```

En promedio la propina es mayor durante las cenas. Hasta el momento podemos decir que si un mesero desea recibir más propina, deberia trabajar durante el servicio de Cena, los domingos y atender preferentemente a hombres

```python
# Afecta al promedio de la propina que el cliente sea fumador o no fumador?
sns.barplot(data=df,x='smoker',y='tip',ci=None)
plt.title ('Propinas promedio Fumador / No Fumador')
plt.xlabel('Fumador/NO Fumador')
plt.ylabel('Propina')
plt.plot()
```

La condición de fumador o de no fumador no afecta a la propina

```python
# Gráfico de barras de propinas promedio por género
sns.barplot(data=df, x='day',y='tip', hue='sex',ci=None)
plt.title('Propinas promedio por día y género')
plt.xlabel('Género')
plt.ylabel('Propina')
plt.show()
```

Podemos analizar este gráfico teniendo en cuenta que perfil del negocio puede ser más rentable dependiendo el día. Aplicacion a negocios, maximización de retornos.

```python
# Gráfico de barras de propinas promedio por tiempo y sexo
sns.barplot(data=df, x='time',y='tip', hue='sex',errorbar=None)
plt.title('Propinas promedio por servicio y género')
plt.xlabel('Género')
plt.ylabel('Propina')
plt.show()
```

```python
# Gráfico de barras de propinas promedio por fumador/no fumador y género
sns.barplot(data=df, x='smoker',y='tip', hue='sex',errorbar=None)
plt.title('Propinas promedio por Fumador/Nofumador y género')
plt.xlabel('Fumador/No Fumador')
plt.ylabel('Propina')
plt.show() 
```

Los hombres fumadore o no fumadores dejan más propina que las mujeres, fumadoras o no

```python
# Gráfico de dispersión entre size y tip
sns.scatterplot(data=df, x='size', y='tip')
plt.title('Relación entre tamaño de de mesa y propinas')
plt.xlabel('Tamaño de mesa')
plt.ylabel('Propina')
plt.show()
```

```python
# Gráfico de barras de propinas promedio según el tamaño de la mesa
sns.barplot(data=df, x='size', y='tip', errorbar=None)
plt.title('Propinas promedio según tamaño de Mesa')
plt.xlabel('Tamaño de mesa')
plt.ylabel('Propina')
plt.show()
```

A mayor cantidad de comensales mayor propina

```python
# Calcular el porcentaje de propina y lo agregamos al DataFrame. Permite hacer unamejor comparación
df['tip_percentage'] = ( df['tip'] / df['total_bill'] ) * 100
df
```

```python
# Gráfico de barras de porcentaje de propinas promedio por sexo
sns.barplot(data=df, x='sex',y='tip_percentage', errorbar=None)
plt.title('Porcentage de propina según Género')
plt.xlabel('Género')
plt.ylabel('Porcentaje de Propina')
plt.show()
```

A pesar de que los hombres dan más dinero (más propina) las mujeres dejan más dinero con respecto al gasto total

```python
#Gráfico de Porcentaje de Propina por día
sns.barplot(data=df, x='day',y='tip_percentage', errorbar=None)
plt.title('Porcentage de propina según el Día')
plt.xlabel('Día')
plt.ylabel('Porcentaje de Propina')
plt.show()
```

Generalmente el porcentaje de propina se mantiene estable, sólo los sábados se puede observar que es un poco menor

```python
sns.barplot(data=df, x='time',y='tip_percentage', errorbar=None)
plt.title('Porcentage de propina según Servicio')
plt.xlabel('Servicio')
plt.ylabel('Porcentaje de Propina')
plt.show()
```

Proporcionalmente en cuanto a porcentaje se ganará una misma cantidad de propina.

```python
sns.barplot(data=df, x='smoker',y='tip_percentage', errorbar=None)
plt.title('Porcentage de propina según Fumador / No Fumador')
plt.xlabel('Fumador/No fumador')
plt.ylabel('Porcentaje de Propina')
plt.show()
```

Podemos observar que no difiere demasiado el porcentaje de propinas teniendo en cuenta si el cliente es o no es fumador
